```python
for element in chemprop['smiles']:
    print(element)
```

```python
smiles_mols = []
for element in range(len(chemprop['smiles'])):
    m = Chem.MolFromSmiles(chemprop['smiles'][element])
    mtemp = Chem.MolToSmiles(m, canonical = True, isomericSmiles = False)
    smiles_mols.append(mtemp)
```

```python
smiles_mols
```

```python
chemprop['smiles_true'] = smiles_mols
```

```python
chemprop
```

```python
chemprop = chemprop.drop(['smiles'], axis=1)
```

```python
chemprop
```

```python
chemprop = chemprop[['smiles_true', 'synergy']]
```

```python
chemprop = chemprop.rename(columns={"smiles_true": "smiles"})
```
